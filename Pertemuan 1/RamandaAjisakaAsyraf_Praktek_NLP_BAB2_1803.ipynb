## Natural Language Toolkit library (NLTK)

merupakan salah satu library yang sangat populer di python untuk pemrosesan bahasa alami.

### Text Processing

dengan nltk.tokenize kita dapat membuat token dari suatu kalimat. contoh penggunaannya dapat dilihat pada kodingan dibawah ini

```python
#!pip install nltk
import nltk
nltk.download('punkt')
text = "Who would have thought that computer programs would be analyzing human sentiments"
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens)
```

### Stopwords

stopwords merupakan kata henti penghubung dari kata tersebut yang tidak memiliki arti. contoh stopword dalam bahasa inggris dapat dilihat dibawah ini.

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
print(stopwords)
```

```python
[word for word in tokens if word not in stopwords]
```

### WordNetlemmatizer

salah satu modul library yang digunakan untuk lematisasi

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
text = "Who would have thought that computer programs would be analyzing human sentiments"
tokens = word_tokenize(text)
lemmatizer = WordNetLemmatizer()
tokens=[lemmatizer.lemmatize(word) for word in tokens]
print(tokens)
```

### PorterStemmer

salah satu modul library untuk melakukan stemming. contoh penggunaanya sebagai berikut

```python
from nltk.stem import PorterStemmer
text = "Who would have thought that computer programs would be analyzing human sentiments"
tokens=word_tokenize(text.lower())
ps = PorterStemmer()
tokens=[ps.stem(word) for word in tokens]
print(tokens)
```

### POS Tag

dalam nltk terdapat pula modul library yang dapat digunakan untuk POS tagging. untuk penggunaanya adalah sebagai berikut

```python
nltk.download('averaged_perceptron_tagger')
nltk.pos_tag(["your"])
nltk.pos_tag(["beautiful"])
nltk.pos_tag(["eat"])
```

```python
text = "Usain Bolt is the fastest runner in the world"
tokens = word_tokenize(text)
[nltk.pos_tag([word]) for word in tokens]
```

## Text Blob

merupakan library yang sangat populer untuk analasis sentimen, penandaan ucapan, terjemahan , dan lainnya. untuk contohnya dapat dilihat sebagai berikut ini

### Sentiment analysis

```python
#!pip install textblob
from textblob import TextBlob
TextBlob("I love pizza").sentiment
TextBlob("The weather is excellent").sentiment
TextBlob("What a terrible thing to say").sentiment
```

### Machine translation

```python
languages = ['fr','zh-CN','hi']
for language in languages:
    print(TextBlob("Who knew translation could be fun").translate(to=language))
```

### Part of speech tagging

```python
TextBlob("The global economy is expected to grow this year").tags
```
